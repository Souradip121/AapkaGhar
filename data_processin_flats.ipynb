```python
def convert_price_to_crore(price_str):
    try:
        # Convert to string and lowercase
        price_str = str(price_str).lower()
        
        # Extract numbers safely
        numbers = re.findall(r'[\d.]+', price_str)
        if not numbers:
            return None
            
        number = float(numbers[0])
        
        # Convert based on unit
        if 'lac' in price_str:
            return round(number / 100,2)
        elif 'crore' in price_str:
            return round(number,2)
        return None
        
    except:
        return None

# Apply conversion
df['price'] = df['price'].apply(convert_price_to_crore)
```

```python
def clean_area_value(area_str):
        # Convert to string
        area_str = str(area_str)
        
        # Remove ₹ and /sq.ft., then strip spaces
        area_flt = float(area_str.replace('₹', '').replace('/sq.ft.', '').replace(',','').strip())
        
        # Convert to numeric (handles commas automatically)
        return area_flt
        
   

# Apply to area column
df['area'] = df['area'].apply(clean_area_value)
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str[0].astype(int)
```

```python
def extract_balcony_number(value):
    try:
        # Convert to string and lowercase
        value = str(value).lower()
        
        # Handle no balcony case
        if 'no balcony' in value:
            return '0'
            
        # Keep original string for 3+ cases
        if '3+' in value:
            return '3+'
            
        # Extract number for standard cases
        number = re.findall(r'\d+', value)
        return number[0] if number else None
        
    except:
        return None

# Apply to column
df['balcony'] = df['balcony'].apply(extract_balcony_number)
```

```python
# Example inputs:
# "2nd of 4 Floors"
# "Ground of 14 Floors"
# "12nd of 14 Floors"

def extract_floor_number(floor_str):
    try:
        # Step 1: Standardize input
        floor_str = str(floor_str).lower()
        # "2nd of 4 Floors" → "2nd of 4 floors"
        
        # Step 2: Handle ground floor
        if 'ground' in floor_str:
            return 0
        
        # Step 3: Extract floor number using regex
        # (\d+) - captures one or more digits
        # (?:st|nd|rd|th)? - optional ordinal suffix
        # \s+of - whitespace followed by "of"
        floor_num = re.search(r'(\d+)(?:st|nd|rd|th)?\s+of', floor_str)
        
        # Step 4: Convert to integer if found
        if floor_num:
            return int(floor_num.group(1))
            
        # Step 5: Return None if no match
        return None
        
    except:
        return None
```

```python
df['floorNum'] = df['floorNum'].apply(extract_floor_number)
```
